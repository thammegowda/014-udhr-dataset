```python
df = align_tsvs(Path('data/tsvs'))
df
```

```python
missing_vals = {'', None, '?', '[missing]', '[?]', '[Missing?]'}
df = df.applymap(lambda x: '' if (x or x.strip()) in missing_vals else x)
df
```

```python

sparse_cols = []
for col_name, n_found in df.applymap(lambda x: bool(x or x.strip())).sum(axis=0).items():
    if n_found <= 50:
        sparse_cols.append(col_name)

print('remove columns', sparse_cols)
len(sparse_cols)
```

```python
skip_cells = []
for col in sparse_cols:
    for lang, val in zip(df['Lang'], df[col]):
        if val:
            skip_cells.append((lang, col, val))

with open('skipped-recs.tsv', 'w') as out:
    for lang, col, val in sorted(skip_cells):
        out.write(f'{lang}\t{col}\t{val}\n')    
```

```python
df_min = df.drop(columns=sparse_cols)
df_min
```

```python
drop_rows = [idx for idx, n_found in df_min.applymap(lambda x: bool(x or x.strip())).sum(axis=1).items() if n_found <= 10]
drop_rows
```

```python
df_min = df_min.drop(index=drop_rows)
```
